```python
import numpy as np
import pandas as pd
```

```python
# Read inputs
isTest = not True
fileName = "test" if isTest else "input"
with open(fileName) as f:
    input = [l.strip("\n") for l in f.readlines()]

print(len(input))
```

```python
def process_sections(lines: [str]):
    def prc_range(lines: [str]):
        def get_range(line: str):
            [des, src, offset] = np.array(line.split()).astype(int)
            return src, des, offset
            # return r, lambda n: n + offset if n in r else n

        ranges = [get_range(l) for l in lines]

        def mapper(src: int):
            for src_org, destination, offset in ranges:
                if src_org <= src < (src_org + offset) :
                    return destination + src - src_org
            return src

        return mapper

    seeds = np.array(lines[0].split(":")[1].split(), dtype=int)
    sts, stf, ftw, wtl, ltt, tth, htl = tuple(
        np.argwhere(np.char.find(input, "map") > 0).T[0]
    )
    return {
        "seeds": seeds.tolist(),
        "seed-to-soil": prc_range(input[sts + 1 : stf - 1]),
        "soil-to-fertilizer": prc_range(input[stf + 1 : ftw - 1]),
        "fertilizer-to-water": prc_range(input[ftw + 1 : wtl - 1]),
        "water-to-light": prc_range(input[wtl + 1 : ltt - 1]),
        "light-to-temperature": prc_range(input[ltt + 1 : tth - 1]),
        "temperature-to-humidity": prc_range(input[tth + 1 : htl - 1]),
        "humidity-to-location": prc_range(input[htl + 1 :]),
    }


map_table = process_sections(input)
map_table.keys()
```

```python
def seed_info(seed: int, mapper: dict):
    def get_destination(src, key: str):
        return mapper[key](src)

    soil = get_destination(seed, "seed-to-soil")
    fertilizer = get_destination(soil, "soil-to-fertilizer")
    water = get_destination(fertilizer, "fertilizer-to-water")
    light = get_destination(water, "water-to-light")
    temperature = get_destination(light, "light-to-temperature")
    humidity = get_destination(temperature, "temperature-to-humidity")
    location = get_destination(humidity, "humidity-to-location")
    return {
        "seed": seed,
        "soil": soil,
        "fertilizer": fertilizer,
        "water": water,
        "light": light,
        "temperature": temperature,
        "humidity": humidity,
        "location": location,
    }


df = pd.DataFrame(list(map(lambda s: seed_info(s, map_table), map_table['seeds'])))
df
```
