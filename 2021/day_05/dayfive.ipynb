```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# filename = "test"
def get_data(filename):
    with open(filename) as f:
        lines = [ln.strip().replace("  ", " ").split(" ") for ln in f.readlines()]

    def get_line(line):
        x1, y1 = line[0].split(",")
        x2, y2 = line[-1].split(",")
        return {"x1": int(x1), "y1": int(y1), "x2": int(x2), "y2": int(y2)}

    return pd.DataFrame([get_line(l) for l in lines])


df = get_data("input")
# dimsize = 10
dimsize = 1000
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
# ax.grid()
mask_vert = df["x1"] == df["x2"]
mask_horz = df["y1"] == df["y2"]
for ix, (x1, y1, x2, y2) in df[mask_horz | mask_vert].iterrows():
    ax = plt.plot([x1, x2], [y1, y2], "-")
```

```python
grid = np.ones((1000, 1000), dtype=np.int16) * -1
for ix, (x1, y1, x2, y2) in df[mask_horz].iterrows():
    # print(" ", x1 - x2, x1, x2)
    for walker_x in range(min(x1, x2), max(x1, x2) + 1):
        # print(walker_x)
        grid[walker_x][y1] += 1

for ix, (x1, y1, x2, y2) in df[mask_vert].iterrows():
    # print(" ", x1 - x2, x1, x2)
    for walker_y in range(min(y1, y2), max(y1, y2) + 1):
        # print(walker_x)
        grid[x1][walker_y] += 1

grid[grid < 0] = 0
print(grid)
# sum(grid)
sum(grid.flatten() > 0)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
ax.grid()
for ix, (x1, y1, x2, y2) in df.iterrows():
    ax = plt.plot([x1, x2], [y1, y2], "-")
```

```python
grid = np.ones((dimsize, dimsize), dtype=np.int16) * -1
for ix, (x1, y1, x2, y2) in df[mask_horz].iterrows():
    # print(" ", x1 - x2, x1, x2)
    for walker_x in range(min(x1, x2), max(x1, x2) + 1):
        # print(walker_x)
        grid[walker_x][y1] += 1

for ix, (x1, y1, x2, y2) in df[mask_vert].iterrows():
    # print(" ", x1 - x2, x1, x2)
    for walker_y in range(min(y1, y2), max(y1, y2) + 1):
        # print(walker_x)
        grid[x1][walker_y] += 1

# Get the Diagonals line or not the lines already y traversed 
for ix, (x1, y1, x2, y2) in df[~mask_vert & ~mask_horz].iterrows():
    heading_x = 1 if x1 < x2 else -1
    heading_y = 1 if y1 < y2 else -1
    for walker_d in range(abs(x1 - x2) + 1):
        walker_x = x1 + walker_d * heading_x
        walker_y = y1 + walker_d * heading_y
        grid[walker_x][walker_y] += 1

grid[grid < 0] = 0
# print(grid)
# sum(grid)
sum(grid.flatten() > 0)
```
